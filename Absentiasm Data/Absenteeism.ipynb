##
1. Dataset Overview

The dataset tracks employee absenteeism at work.

Each row represents one recorded absence, with details on:

Date of absence

Reason for absence (coded categories)

Work-related and personal factors such as seasons, distance from home, service time, age, daily workload, education, and health-related metrics.

Target column: Absenteeism time in hours.


2. Cleaning Performed

Standardized column names for consistency.

Converted the Date column into datetime format for easier time-based analysis.

Added a derived column for day of the week to identify weekday trends.

Checked and handled missing values — removed or imputed where appropriate.

Ensured categorical columns (e.g., Reason for Absence) have consistent category codes.

Removed duplicate rows to avoid skewing analysis.

3. Key Observations from the Data

Absenteeism reasons are diverse — top categories might be medical issues, personal reasons, or work-related factors (need deeper breakdown).

Data covers multiple seasons, allowing for seasonal absenteeism trend analysis.

Age, service time, and education level are recorded — enabling demographic comparisons.

Distance from home and daily workload could indicate commute stress or work pressure links to absenteeism.


4. How This Clean Data Can Be Used

Trend analysis: Examine which days of the week see the most absences.

Seasonal patterns: Identify if absences spike in certain months or weather conditions.

Reason distribution: Determine the most common causes of absence.

Predictive modeling: Build a model to predict absenteeism risk based on employee attributes.

HR decision support: Use insights to create policies that reduce avoidable absenteeism
